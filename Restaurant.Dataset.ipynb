```python
#5.Rename columns
sidd=sidd.rename(columns={"sex":"Gender"})
sidd=sidd.rename(columns={"total_bill":"Total_bill"})
sidd=sidd.rename(columns={"tip":"Tip"})
sidd=sidd.rename(columns={"smoker":"Smoker"})
sidd=sidd.rename(columns={"day":"Day"})
sidd=sidd.rename(columns={"time":"Time"})
sidd=sidd.rename(columns={"	size":"	Size"})
print(sidd)
```

```python
# 6.Datatype conversion
sidd.dtypes
```

```python
# 7.String operator
sidd["Gender"]=(sidd["Gender"]
            .str.replace("Female","F")
            .str.replace("Male","M"))
print(sidd["Gender"].to_string())
```

```python
sidd
print(sidd["Day"].unique())
```

```python
sidd["Day"] = sidd["Day"].astype(str)

day_map = {
    "Sun": "Sunday",
    "Sat": "Saturday",
    "Thur": "Thursday",
    "Fri": "Friday"
}

sidd["Day"] = sidd["Day"].replace(day_map)

print(sidd["Day"])
```

```python
# Machine Learning
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

sidd["Gender_encode"]=encoder.fit_transform(sidd["Gender"])
sidd["Smoker_encode"]=encoder.fit_transform(sidd["Smoker"])
sidd[["Smoker","Smoker_encode","Gender","Gender_encode"]]
```

```python
import pandas as pd

mugu= pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female"],
    "Purchased": ["Yes", "No", "Yes", "No", "Yes", "No"]
})
mugu["Gender_encode"]=encoder.fit_transform(mugu["Gender"])
mugu["Purchased_encode"]=encoder.fit_transform(mugu["Purchased"])
sidd["Smoker_encode"]=encoder.fit_transform(sidd["Smoker"])

mugu[["Gender","Gender_encode"]]
mugu[["Purchased","Purchased_encode"]]
sidd[["Smoker","Gender_encode"]]

print(mugu)
```

```python
mugu_day=pd.get_dummies(
    sidd,
    columns = ["Day"],
    dtype=int
)
print(mugu_day)
```

```python
day_mean=sidd.groupby("Day")["Tip"].mean()
print(day_mean)
sidd["day_encoded"]=sidd["Day"].map(day_mean)
sidd[["Day","day_encoded"]]
```

```python
from sklearn.preprocessing import StandardScaler

num_colm=["Total_bill","Tip","size"]
scaler=StandardScaler()

sidd[num_colm]=scaler.fit_transform(sidd[num_colm])

print(sidd.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

num_colm=["Total_bill","Tip","size"]
scaler=MinMaxScaler()

sidd[num_colm]=scaler.fit_transform(sidd[num_colm])

print(sidd.head())
```

```python
from sklearn.preprocessing import RobustScaler

num_colm=["Total_bill","Tip","size"]
scaler=RobustScaler()

sidd[num_colm]=scaler.fit_transform(sidd[num_colm])

print(sidd.head())
```
